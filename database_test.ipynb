```python
import pandas as pd

csv_file = '2024_cluster3_predictions.csv'

df = pd.read_csv(csv_file)
df.rename(columns={df.columns[0]: 'prediction'}, inplace=True)

# Generate a new column with dates dividing the year into 24 periods
df['date'] = pd.date_range(start='1/1/2024', end='31/12/2024', periods=24)

# Change the first value of the prediction column to 0
df.at[0, 'prediction'] = 60000

print(df)
```

```python
# df to sql
import sqlite3
conn = sqlite3.connect('predictions.db')

df.to_sql('predictions', conn, if_exists='replace', index=False)
```

```python
# read from sql
df = pd.read_sql_query("SELECT * FROM predictions", conn)
print(df)
```
